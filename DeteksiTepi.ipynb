## import libary

```python
import cv2 as marcel
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data 
from skimage.io import imread
from skimage.color import rgb2gray 
```

## Deteksi Gambar

```python
pict = marcel.imread('parkir.jpeg')
```

```python
marcel.imshow('pict asli', pict)
marcel.waitKey(0)
marcel.destroyAllWindows()
```

## Show Tepi gambar
#fungsi edges adalah untuk mendeteksi gambar

```python
pict1 = marcel.cvtColor(pict, marcel.COLOR_BGR2GRAY)
edges = marcel.Canny(pict, 100, 150) 
```

## Showing with figure and axis (matplotlib function)

```python
fig, axs = plt.subplots(1,2, figsize = (10,10))
ax = axs.ravel()

ax[0].imshow(pict1,cmap = 'gray')
ax[0].set_title ('pict ori')

ax[1].imshow(edges,cmap = 'gray')
ax[1].set_title ('setelah di edges')
```

## Showing edge on Image
##lines = marcel.HoughLinesP(edges, 1, np.pi/180, 30, maxLineGap=250): Pada bagian ini, kita menggunakan metode HoughLinesP() dari objek atau modul yang disebut marcel. Metode ini digunakan untuk mendeteksi garis-garis dalam gambar menggunakan Transformasi Hough Probabilistik. Argumen yang diberikan adalah:

edges: Citra tepi (citra biner) yang digunakan sebagai input untuk mendeteksi garis.
1: Resolusi jarak dalam piksel.
np.pi/180: Resolusi sudut dalam radian.
30: Threshold, yaitu jumlah minimum pemilih yang diperlukan untuk menerima garis.
maxLineGap=250: Jarak maksimum antara segmen garis yang masih dianggap sebagai garis tunggal.
Hasil dari metode HoughLinesP() disimpan dalam variabel lines.

```python
lines = marcel.HoughLinesP(edges, 1, np.pi/180, 30, maxLineGap=250)
pict_line = pict.copy()
```

```python
for line in lines:
    x1, y1, x2, y2 = line[0]
    marcel.line(pict_line, (x1, y1), (x2, y2), (0, 0, 225), 1)
```

## Showing With Figure & Axis (matplotlib Function)

```python
pict2 = marcel.cvtColor(pict_line, marcel.COLOR_BGR2RGB)

fig, axs = plt.subplots(1,2, figsize = (10,10))
ax = axs.ravel()

ax[0].imshow(pict1,cmap = 'gray')
ax[0].set_title ('pict ori')

ax[1].imshow(pict2,cmap = 'gray')
ax[1].set_title ('setelah di edges')
```
