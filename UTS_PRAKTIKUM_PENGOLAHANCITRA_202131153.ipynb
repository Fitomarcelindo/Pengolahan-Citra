# B. Lakukan Convert GRAYSCALE TO BINARY

```python
(thresh, binary) = cv.threshold(img_gray, 128, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)
```

```python
cv.imshow("GrayScale to Binary", binary)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
cv.imwrite("GrayScale Image.jpg", binary)
```

# 2. FILTER KONVOLUSI

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
citra = imread(fname= "Bahan/2.png")

print(citra.shape)

plt.imshow(citra, cmap='gray')
```

```python
kernel = np.array ([[-1, 0, -1],
                   [0, 4, 0],
                   [-1, 0, -1]])

citraOutput = cv2.filter2D(citra, -1, kernel)

flig, axes = plt.subplots(1,2, figsize=(15,11))
ax = axes.ravel()

ax[0].imshow(citra, cmap='gray')
ax[0].set_title("Citra Input")
ax[1].imshow(citraOutput, cmap='gray')
ax[1].set_title("Citra Output")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Bahan/2.png')
[baris, kolom] = img.shape[:2]

# Meningkatkan kecerahan citra dengan menambahkan beta
beta = 40
citra_cerah = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = img[x, y] + beta
        citra_cerah[x, y] = gyx

citra_cerah = citra_cerah.astype(np.uint8)

# Meregangkan kontras citra dengan mengalikan alpha
alpha = 1.2
citra_kontras = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = citra_cerah[x, y] * alpha
        citra_kontras[x, y] = gyx

citra_kontras = citra_kontras.astype(np.uint8)

# Membalik citra
inverted_img = np.fliplr(citra_kontras)

# Plotting hasil
fig, axs = plt.subplots(2, 2, figsize=(15, 5))
axs[0, 0].imshow(img)
axs[0, 0].set_title("Citra Asli")

axs[0, 1].hist(img.ravel(), 256, [0, 256], color='red')

axs[1, 0].imshow(inverted_img)
axs[1, 0].set_title("Citra hasil meningkatkan kecerahan, meregangkan kontras, dan membalik citra")

axs[1, 1].hist(inverted_img.ravel(), 256, [0, 256], color='#065535')

plt.show()
```
