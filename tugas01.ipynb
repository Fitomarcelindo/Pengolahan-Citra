# Brightness

```python
beta = 33 #menggunakan nilai konstanta beta 33 bias, range brightness 0-100
img_cerah = np.zeros((baris, kolom, 3))#pada baris ini membuat variabel baru dengan baris dan kolom 3 
for x in range (baris): #perulangan ini melakukan loop untuk setiap 
    #pixel 
    for y in range (kolom):
        gyx = img1[x,y]+beta #penambahaan konstanta beta
        img_cerah [x,y] = gyx
        
img_cerah = img_cerah.astype(np.uint8)A#varaiabel img_cerah menjadi 8 bit karena fungsi astype range nilai kecerahaan harus 0-255 

#sebelum
fig, axs = plt.subplots(2,2, figsize =(21, 7))#membuat 2 abris 2 kolom dengan size 22*7 inch
axs[0,0].imshow(img1)
axs[0,1].hist(img1.ravel(), 256, [0,256], color='gray')

#sesudah
axs[1,0].imshow(img_cerah)
axs[1,1].hist(img1.ravel(), 256, [0,256], color='green')
plt.show() #menampilakan keselururhaan program yang dituliskan FittoMartcellindo_20213001s
```

# Contras


```python
alpa = 33
img_contras = np.zeros((baris, kolom, 3))

for x in range (baris):
    for y in range (kolom):
        
        gyx = img1[x,y]*alpa
        img_contras [x,y] = gyx
img_contras = img_contras.astype(np.uint8)

#sebelum
fig, axs = plt.subplots(2,2, figsize =(21, 7))
axs[0,0].imshow(img1)
axs[0,1].hist(img1.ravel(), 256, [0,256], color='gray')

#sesudah

axs[1,0].imshow(img_contras)
axs[1,1].hist(img1.ravel(), 256, [0,256], color='green')
plt.show()
```
