```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
citra = imread(fname="noisyimage.jpg")

print(citra.shape)

plt.imshow(citra, cmap='gray')
```

```python
kernel = np.array([[-1, 0, -1],
                  [0, 4, 0],
                  [-1, 0, -1]])
citraOutput = cv2.filter2D(citra, -1, kernel)

flig, axes =plt.subplots(1,2, figsize=(15,11))
ax =axes.ravel()

ax[0].imshow(citra, cmap='gray')
ax[0].set_title("Citra Input")
ax[1].imshow(citraOutput, cmap='gray')
ax[1].set_title("Citra Output")
```

```python
#Histogram
img = cv2.imread('danautoba.jpg')
[baris, kolom] = img.shape[:2]

hist = cv2.calcHist([img],[0],None,[256],[0,256])
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))

axs[0].imshow(img)
axs[1].plot(hist)
plt.show()
```

```python
img =cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
fig, axs =plt.subplots(1,2,figsize=(15,5))

axs[0].imshow(img)
axs[1].plot(hist, color='g')
plt.show()
```

```python
fig, axs =plt.subplots(2,2,figsize=(15,5))

axs[0,0].imshow(img)
axs[0,1].plot(hist, color='g')

axs[1,0].imshow(img)
axs[1,1].hist(img.ravel(),256,[0,256],color='r')
plt.show()
```

```python
img_abu = cv2.imread('danautoba.jpg',0)
hist2=cv2.equalizeHist(img_abu)

fig, axs =plt.subplots(1,2,figsize=(15,5))

axs[0].imshow(cv2.cvtColor(img_abu, cv2.COLOR_BGR2RGB))
axs[1].hist(img_abu.ravel(),256,[0,256])
plt.show()
```

```python
beta = 40
citra_cerah = np.zeros((baris,kolom, 3))

for x in range(baris) :
    for y in range(kolom) :
        gyx = img[x,y] + beta
        citra_cerah[x,y] = gyx
        
citra_cerah = citra_cerah.astype(np.uint8)

fig, axs =plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(),256,[0,256],color ='red')

axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.ravel(),256,[0,256],color= '#065535')
plt.show()
```

```python
beta = 10
alpa = 1.2
img_hasil = np.zeros((baris, kolom, 3))
for x in range(baris):
    for y in range(kolom):
        gyx = alpa * img[x,y] + beta
        img_hasil[x,y] = gyx

img_hasil=img_hasil.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(),256,[0,256])

axs[1,0].imshow(img_hasil)
axs[1,1].hist(img_hasil.ravel(),256,[0,256])
plt.show() 
```

```python
alpa = 1.2
citra_kontras = np.zeros((baris, kolom, 3))
for x in range(baris):
    for y in range(kolom):
        gyx = img[x,y] * alpa
        citra_kontras[x,y] = gyx
        
citra_kontras = citra_kontras.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize = (15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(),256,[0,256])

axs[1,0].imshow(citra_kontras)
axs[1,1].hist(citra_kontras.ravel(),256,[0,256])
plt.show()
```
