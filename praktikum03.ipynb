```python
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
import cv2

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
img = imread(fname = "noisyimage.jpg")
print(img.shape)
plt.imshow(img,cmap = 'gray')
```

```python
kernel = np.array([[-1, 0, -1],
                  [0, 4.3, 0],
                  [-1, 0, -1]])
citraOutput = cv2.filter2D(img, -1, kernel)

fig, axes = plt.subplots(1,2, figsize=(12,12))
ax = axes.ravel()

ax[0].imshow(img, cmap = 'gray')
ax[0].set_title('gamabrfito')
ax[0].imshow(citraOutput, cmap = 'gray' )
ax[1].set_title('gambarfito')
```

# meningkatkan kecerahaan 

```python
beta = 33
img_cerah = np.zeros((baris, kolom, 3))

for x in range (baris):
    for y in range (kolom):
        gyx = img1[x,y]+beta
        img_cerah [x,y] = gyx
img_cerah = img_cerah.astype(np.uint8)

#sebelum
fig, axs = plt.subplots(2,2, figsize =(21, 7))
axs[0,0].imshow(img1)
axs[0,1].hist(img1.ravel(), 256, [0,256], color='gray')

#sesudah

axs[1,0].imshow(img_cerah)
axs[1,1].hist(img1.ravel(), 256, [0,256], color='green')
plt.show()
```
