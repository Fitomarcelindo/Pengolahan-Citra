# citra one 


```python
import cv2 as marcel
import numpy as np
```

```python
citra =  marcel.imread('noisyimage.jpg')
```

```python
marcel.imshow("citra.asli", citra)
marcel.waitKey(0)
marcel.destroyAllWindows()
```

```python
print(citra.shape)
print(citra.dtype)
```

```python
b = citra[:,:,0]
g = citra[:,:,1]
r = citra[:,:,2]
```

```python
panjang = len(citra)
lebar = len(citra[0])

citra_gray = np.zeros((panjang,lebar))

for i in range (panjang):
    for j in range(lebar):
        citra_gray[i,j]=0.2989 * r[i,j]+0.587* g[i,j]+0.1141* b[i,j]
citra_gray=citra_gray.astype(np.uint8)
```

```python
marcel.imshow("Abu", citra_gray)
marcel.waitKey(0)
marcel.destroyAllWindows()
```

```python
print(citra_gray)
```

## grayscale to binary

```python
(thresh, binary) = marcel.threshold(citra_gray, 128, 256, marcel.THRESH_BINARY | marcel.THRESH_OTSU)
```

```python
marcel.imshow(" grey scale binary", binary)
marcel.waitKey(0)
marcel.destroyAllWindows()
```

## citra konlousi

```python
import cv2 as marcelf
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
citra = imread(fname="danautoba.jpg")

print(citra.shape)

plt.imshow(citra,cmap='gray')
```

```python
kernel = np.array([[-1,0,-1],
                  [0,4,0],
                  [-1,0,-1]])

citraOutput = marcelf.filter2D(citra,-1,kernel)

fig,axes = plt.subplots(1,2, figsize=(15,11))
ax = axes.ravel()

ax[0].imshow(citra, cmap='gray')
ax[0].set_title("Citra Input")
ax[1].imshow(citraOutput, cmap='gray')
ax[1].set_title("Citra Input")
```

# histogram

```python
img = marcel.imread('danautoba.jpg')
[baris, kolom] = img.shape[:2]

hist= cv2.calcHist([img],[0],None,[256],[0,256])
```

```python
alpa = 1.2
citra_kontras = np.zeros((baris, kolom, 3))
for x in range(baris):
    for y in range(kolom):
        gyx = img[x,y] * alpa
        citra_kontras[x,y] = gyx
        
citra_kontras = citra_kontras.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(),256,[0,256])

axs[1,0].imshow(citra_kontras)
axs[1,1].hist(citra_kontras.ravel(),256,[0,256])
plt.show()
```

```python
beta = 40
citra_cerah = np.zeros((baris, kolom, 3))

for x in range(baris) : 
    for y in range(kolom) : 
        gyx = img[x,y] + beta
        citra_cerah[x,y] = gyx
        
citra_cerah = citra_cerah.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(),256,[0,256], color = 'red')

axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.ravel(),256,[0,256], color = '#065535')
plt.show()
```
