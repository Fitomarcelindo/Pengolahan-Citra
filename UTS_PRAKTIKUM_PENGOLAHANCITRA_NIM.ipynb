# 1. CONVERT CITRA 

# A. RBG TO GRAYSCALE

```python
import cv2
import numpy as np
```

```python
citra = cv2.imread('rgb.jpg')
```

```python
cv2.imshow("citra.asli", citra)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(citra.shape)
print(citra.dtype)
```

```python
b = citra[:,:,0]
g = citra[:,:,1]
r = citra[:,:,2]
```

```python
panjang = len(citra)
lebar = len(citra[0])

citra_gray = np.zeros((panjang,lebar))

for i in range (panjang):
    for j in range(lebar):
        citra_gray[i,j]=0.2989 * r[i,j]+0.587* g[i,j]+0.1141* b[i,j]
citra_gray=citra_gray.astype(np.uint8)
```

```python
cv2.imshow("Abu", citra_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(citra_gray)
```

# B. GRAYSCALE TO BINARY

```python
panjang = len(citra)
lebar = len(citra[0])

ambang = 210

citra_biner = np.zeros((panjang,lebar))

for i in range (panjang):
    for j in range (lebar):
        if citra_gray[i,j] >= ambang:
          citra_biner[i,j] = 0
        else:
          citra_biner[i,j] = 1
```

```python
cv2.imshow("biner", citra_biner)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 2. FILTER KONVOLUSI

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
citra = imread(fname="1.jpg")

print(citra.shape)

plt.imshow(citra,cmap='gray')
```

```python
kernel = np.array([[-1,0,-1],
                  [0,4,0],
                  [-1,0,-1]])

citraOutput = cv2.filter2D(citra,-1,kernel)

fig,axes = plt.subplots(1,2, figsize=(15,11))
ax = axes.ravel()

ax[0].imshow(citra, cmap='gray')
ax[0].set_title("Citra Input")
ax[1].imshow(citraOutput, cmap='gray')
ax[1].set_title("Citra Input")

```

# 3.Histogram

```python
#Histogram
img = cv2.imread('1.jpg')
[baris, kolom] = img.shape[:2]

hist= cv2.calcHist([img],[0],None,[256],[0,256])
```

```python
#Merenggangkan Kontras
alpa = 1.2
citra_kontras = np.zeros((baris, kolom, 3))
for x in range(baris):
    for y in range(kolom):
        gyx = img[x,y] * alpa
        citra_kontras[x,y] = gyx
        
citra_kontras = citra_kontras.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(),256,[0,256])

axs[1,0].imshow(citra_kontras)
axs[1,1].hist(citra_kontras.ravel(),256,[0,256])
plt.show()
```

```python
#Meningkatkan kecerahan
beta = 40
citra_cerah = np.zeros((baris, kolom, 3))

for x in range(baris) : 
    for y in range(kolom) : 
        gyx = img[x,y] + beta
        citra_cerah[x,y] = gyx
        
citra_cerah = citra_cerah.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(),256,[0,256], color = 'red')

axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.ravel(),256,[0,256], color = '#065535')
plt.show()
```

```python
#MembalikCitra
import cv2
import matplotlib.pyplot as plt

# Membaca citra
img = cv2.imread('1.jpg')

# Membalik citra
inverted_img = cv2.bitwise_not(img)

# Menampilkan citra asli dan citra yang sudah dibalik
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0].set_title('Citra Asli')

axs[1].imshow(cv2.cvtColor(inverted_img, cv2.COLOR_BGR2RGB))
axs[1].set_title('Citra yang Dibalik')

plt.show()
```
